## 4- DATABASE/SQL:
Given the two database tables below:
-  i_users (userId, username, email address, isActive, gender) stores the users.
- i_user_login_logs (userId, login_date) stores the multiple login dates of the users in i_users table.

Please write an SQL query to determine how many different users logged in more than three times.
(preferred to have your SQL query pasted here instead of sharing as a link) 

**Note: The output should return a single row including the number of users as requested.**


### answer:
SELECT  COUNT(userId)

FROM (SELECT userId,COUNT(login_date) AS login_count FROM i_user_login_logs GROUP BY userId) AS a

WHERE login_count>3

Please write an SQL query to determine the username and email address of the user(s) who logged in most (administrator user should be excluded).

### answer:

FROM i_users AS i

LEFT JOIN

(SELECT userId,COUNT(login_date) AS logincount FROM i_user_login_logs GROUP BY userId) AS p

ON i.userId=p.userId

## 8- TELECOM
Please describe the differences between 3G and 4G technologies and what will change once 5G comes to play.
### answer:
There are some differences between 3G and 4G technologie Like:
1.	Maximum upload rate of 3G technology is 5 Mbps.	While the maximum upload rate of 4G technology is 500 Mbps.
2.	Maximum download rate of 3G technology is 21 Mbps.	While the maximum download rate of 4G technology is 1 Gbps.
3.	3G uses packet switching technique.	While 4G uses packet switching technique as well as message switching technique.
4.	The frequency range of 3G technology is from 1.8 GHz to 2.5 GHz.While 4G’s frequency range is from 2 GHz to 8 GHz.
5.	3G lenient horizontally. While 4G lenient horizontally as well as vertically.
6.	3G is a wide area cell based network architecture. While 4G is the integration of Wireless LAN as well as Wide Area cell based network architecture.
7.	There is turbo codes are used for error correction in 3G technology. 4G technology uses concatenated codes for error correction

One of the biggest differences between 4G and 5G will be **peak capacity and latency**.

For example, peak capacity of 5G UWB sector is in gbps compared to 4G in mbps. Also, the latency, or the time that passes from the moment information is sent from a device until it is used by a receiver, will be greatly reduced on 5G networks, allowing for faster upload and download speeds.

Another big difference between 4G and 5G is **bandwidth size**. 5G should be able to support many more devices of the future, in addition to the network demands of connected vehicles and other devices in the Internet of Things
