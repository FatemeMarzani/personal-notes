# Week 1 - Introducing NoSQL

- NoSQL means **Not only SQL.** 

- NoSQL databases have their roots in the open source community. 

- NoSQL database **implementations are technically different**  from each other. 

## The four main categories of NoSQL database 
- **Key-Value:** Key-Value NoSQL databases are the least complex architecturally.
- **Document:** Document-based NoSQL databases use documents to make values visible for queries. In document-based NoSQL databases, each piece of data is considered a document, which is typically stored in either JSON or XML format.
- **Wide Column:** The primary use cases for column-based NoSQL databases are event logging and blogs, counters, and data with expiration values.
- **Graph:** Graph databases store information in entities (or nodes) and relationships (or edges)

## ACID principles
**ACID** stands for **A**tomicity, **C**onsistency, **I**solated, **D**urable

## BASE principles
**BASE** stands for **B**asic **A**vailability, **S**oft-state, **E**ventual Consistency. BASE stands for Basic Availability, Soft-state, Eventual Consistency. 

## CAP Theorem
These are Consistency, Availability, and Partition Tolerance, or CAP. A distributed system can guarantee delivery of only two of these three desired characteristics.

- You can use the CAP Theorem to classify NoSQL databases.

- **Partition Tolerance is a basic feature of NoSQL databases**

## RDBMS or NoSQL?
- NoSQL systems are not a de facto replacement of RDBMS. 

- RDBMS and NoSQL cater to different use cases, which means that your solution could use both RDBMS and NoSQL.
